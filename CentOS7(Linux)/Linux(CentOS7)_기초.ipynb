## Linux 개념

Linux는 오픈소스 운영체제(OS)이다. 운영체제는 CPU, 메모리, 스토리지처럼 시스템의 하드웨어와 리소스를 직접 관리하는 소프트웨어이다. OS는 애플리케이션과 하드웨어사이에 모든 소프트웨어와 작업을 수행하는 물리적 리소스를 연결한다.

## Linux 명령어

컴퓨터 이름 호출

```
hostname
```

인터넷 연결 확인

```
ping 8.8.8.8 # 지정한 ip는 구글
```

아이피 확인

```
ipconfig
```

아이피만 확인

```
hostname -i
```

내용 지우기

```
clear
Ctrl + L
```

재시작

```
restart
```

구성 파일 적용

```
reload
```

설치된 패키지 업데이트

```
yum update - y
```

관리자 권한으로 명령(sudo)

```
sudo yum update -y # 관리자 권한 업데이트
sudo reboot # 관리자 권한 재부팅
```

사용한 명령어 확인

```
history
```

history를 파일로 저장

```
history > mycommand.list
```

[Doker](https://www.redhat.com/ko/topics/containers/what-is-docker) 설치

```
curl -sSL http://get.docker.com | sh
```

Doker 실행

```
systemctl start docker
docker pull centos
docker run =it centos /bin/bash
```

용량 확인

```
df
df -h # 용량을 단위로 표시
```

접속 계정 바꾸기

```
su [계정 이름]
```

현재 계정 다시 로그인

```
su - # 기존의 설정된 계정 로그인
su - {계정} # 입력한 계정 로그인
```

계정 비밀번호 재설정

```
passwd # 접속 중인 계정 비번 재 설정
passwd {계정} # 지정한 계정 비번 재 설정
```

사용자 생성

```
useradd {유저명}
```

접속 계정 확인

```
whoami
```

계정 접속 시간

```
who
```

최초 계정 접속자(하루 기준)

```
who am i
```

글자를 찾고 찾은 행 전체 반환

```
grep
```

홈(기본) 경로 이동

```
cd ~
```

다음 경로 

```
cd {이동할 경로}
```

이전 경로

```
cd - # 이전 경로 이동 하면서 설정된 경로 표시
cd .. # 이전 경로 이동
```

루트 경로

```
cd /
```
현재 디렉토리 확인

```
pwd
```

디렉토리 확인

```
ls
ls -al # 디렉토리를 상세히 확인 가능
```

홈 디렉토리 확인

```
ls-| ~{계정} # root 상태에서 확인 가능
```

파일 종류 확인

```
file 파일명
```

디렉토리 만들기

```
mkdir {디렉토리명} # 현재 위치에 생성
```

디렉토리 삭제

```
rmdir {디렉토리명} 
```

명령어 구분

```
who;whoami # who, whoami 명령어 각각 실행
```

Terminal 종료

```
exit
Ctrl + d
```
