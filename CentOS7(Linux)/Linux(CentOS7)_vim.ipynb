## vim단축키

윈도우 체제에서 Linux, Ubuntu 넘어 오게되면 콘솔창을 많이 사용하게 된다.<br>
콘솔창에 익숙해 지기 위해서 vim단축키에 대해 알아보자.

**시작 명령어**

vi {파일명}.파일형식 - 파일열기, 작성<br>
vi -행번호 {파일명}.파일형식 - 파일 열고, 입력한 행으로 커서 이동<br>
vi -r {파일명}.파일형식 - 손상된 파일 회복<br>
view {파일명}.파일형식 - 읽기 전용으로 열기

**방향**

H - 왼쪽<br>
J - 아래쪽<br>
K - 위쪽<br>
L - 오른쪽

**수정**

i - 현재 커서 위치 insert<br>
Shift + i - 현재 줄 맨 앞쪽 insert<br>
a - 현재 커서 뒤쪽 insert<br>
Shift + a - 핸재 줄 맨 뒤쪽 insert<br>
o - 아랫 줄에 insert<br>
Shift + o - 윗 줄에 insert<br>
dd - 현재 줄 잘라내기<br>
dw - 커서 위치 단어 잘라내기<br>
Shift + d - 현재 커서 위치 부터 줄의 끝까지 잘라내기<br>
:5,10d - 5_10번째 행 삭제<br>
s - 커서에 위치한 문자 잘라내고 insert<br>
Shift + s - 커서가 위치한 줄 잘라내고 insert<br>
r - 현재 글자 교체하기<br>
r + Enter - 행 분리
u - Undo(이전 단계)<br>
Shift + r : Redo(이후 단계)<br>
x - 현재 글자 지우기<br>
Shift + x - 앞의 글자 지우기<br>
〉 - 들여쓰기<br>

〈 - 내어쓰기<br>

yy - 현재 줄 복사<br>
yw - 현재 커서의 위치부터 단어가 끝나는 위치까지 복사<br>
yl - 커서에 위치한 문자 복사<br>
yh - 커서의 왼쪽 문자 복사<br>

yk - 커서가 위치한 줄과 그 윗줄 복사<br>
p - 단어 복사 현재 커서가 위치한 바로 다음 위치에 붙여넣기, 행 복사 현재 커서가 위치한 줄 바로 아랫줄에 붙여넣기<br>
Shift + p - 단어 복사 현재 커서가 위치한 바로 앞 위치에 붙여넣기, 행 복사 현재 커서가 위치한 줄 바로 윗줄에 붙여넣기<br>
숫자 + p - 해당 숫자 만큼 붙여넣기<br>
:1,2co 3 - 1_2행을 3행 다음으로 붙여넣기<br>

## gcloud 연동(gcloud SDK)

GCP에 자료를 원활히 업로드하기 위해 SDK를 활용하여 연동이 필요<br>
우선 gcloud SDK 저장소 정보로 yum으로 업데이트 진행

```
sudo tee -a /etc/yum.repos.d/google-cloud-sdk.repo << EOM
[google-cloud-sdk]
name=Google Cloud SDK
baseurl=https://packages.cloud.google.com/yum/repos/cloud-sdk-el7-x86_64
enabled=1
gpgcheck=1
repo_gpgcheck=1
gpgkey=https://packages.cloud.google.com/yum/doc/yum-key.gpg
       https://packages.cloud.google.com/yum/doc/rpm-package-key.gpg
EOM
```

gcloud SDK.PNG<br>
암호 입력창에 CentOS 암호 입력<br>

gcloud 설치를 진행, install 명령을 실행 할때는 root(관리자)모드로 진입하여 실행<br>
root모드로 진입하기 위해서 ```su -```를 입력한 뒤 CentOS암호 입력<br>
root모드로 진입 했으면 아래의 코드를 실행하여 install 명령을 실행

```
yum install google-cloud-sdk
```

위의 작업을 실행하면 중간에 패키를 설치 여부를 물어봄, 설치를 권장<br>
아래의 명령어를 사용하여 SDK 초기화 작업을 진행하고 구글 계정에 연동을 진행하자!

```
gcloud init
```

위의 명령어를 입력하면 ```You must log in to continue. Would you like to log in (Y/n)?``` 창에서 ```Y```를 입력(구글 사용자 계정으로 로그인 한다는 동의 문구)<br>
Y를 클릭하면 구글창이 생성, 접속할 아이디 입력하고 동의를 누르면 동의한 아이디의 프로젝트명이 출력

```
Pick cloud project to use:
 [1] [my-project-1]
 [2] [my-project-2]
 ...
 Please enter your numeric choice:
 ```

 위와 같은 형식으로 생성되는데 연동할 프로젝트를 번호로 지정하여 설정<br>
아래의 코드를 이용하여 연동이 잘 됐는지 확인해보자!

```
gcloud info
```

위의 코드를 이용하면 gcloud에 연동된 목록을 확인 할수 있음




## error


### 1.계정에 관한 에러

CentOS에서 yum 명령어로 install할때 아래의 에러가 발생

```
loaded plugins: fastesmirror, langpacks you need be root to perform this command
```

위의 에러는 관리자(root) 계정이 아닌 사용자 계정으로 yum을 실행하려고 해서 발생<br> 해결 방법은 명령창에 아래와 같이 입력하여 관리자(root) 모드로 전환

```
# su -<br>
암호: {관리자 비밀번호 입력}
```


### 2.인터넷 연결 문제로 인한 에러

vim 설치하기 위해 yum 명령어를 사용하여 인스톨을 실행시 아래의 에러가 발생

```
loaded plugins: fastestmirror
```

위의 에러는 인터넷이 연결되어 있지 않아서 발생하는 경우가 많음<br>
아래와 같이 입력해서 인터넷이 설정 되어 있는지 확인

```
vi /etc/sysconfig/network-scripts/ifcfg-enp0s3
```

위의 명령어를 실행하면 아래와 같은 창이 생성


CentOS인터넷1.PNG<br>
ONBOOT가 no 설정되어 인터넷 연결이 안됨<br>
아래와 같이 설정을 수정

CentOS인터넷2.PNG<br>
변경 하고자 하는 곳으로 이동 한뒤 insert키 ```i```를 눌러 yes로 수정<br>
수정을 한뒤 esc를 누르고 ```:wq```를 입력하고 나오자.

변경한 값이 설정 되도록 아래와 같이 입력하여 재부팅

```
# reboot
```

인터넷이 원활하게 연결되는지 확인하기 위해 테스트를 진행 (8.8.8.8은 구글의 주소 값)

```
# ping 8.8.8.8 
```

원활하게 연결 된다면 패킷을 내보 내는 것을 확인 할 수 있음<br>
종료를 하기 위해서는 Ctrl + C를 클릭하여 종료

### 3. gem 명령 오류(vimgolf설치)

CentOS7으로 vimgolf를 설치하기 위해 아래와 같은 명령을 실행

```
$ gem install vimgolf
```

위의 코드를 실행하자 ```bash: gem: 명령을 찾을 수 없습니다...``` 문구와 함께 명령이 실행 되지 않음<br>
이건 현재 환경에서 Rubygems가 설치 되어 있지 않아서 발생하는 오류<br>
아래의 코드를 입력하여 Rubygems을 설치(위: Linux 명령어, 아래: Ubuntu 명령어)

```
sudo yum install rubygems
sudo apt-get install rubygems
```

입력 후 실행 시 아래와 같이 실행

Ruby.PNG<br>
위의 Cento의 암호에 나의 Cento비번 입력

ruby in.PNG<br>
설치가 완료 되면 vimgolf설치 인스톨을 실행하여 설치

in.PNG<br>
위의 창과 같이 뜬다면 정상적으로 설치

### 4. 저장소 에러(EPEL)

centOS7에서 인스톨 설치를 위해 yum코드를 사용하여 설치를 하는 중 아래와 같은 에러가 발생

```
Error getting repository data for epel, repository not found
```

위의 에러는 repo(저장소)를 찾지 못한다는 에러로 저장소를 설치하면 해결<br>
아래의 코드를 실행하여 저장소를 다시 설치

```
rpm -Uvh https://dl.fedoraproject.org/pub/epel/epel-release-latest-7.noarch.rpm
```

저장소가 새로 생성되면서 에러가 해결
